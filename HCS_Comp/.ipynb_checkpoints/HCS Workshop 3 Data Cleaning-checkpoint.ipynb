```python
cardboard_str='archive/Garbage classification/Garbage classification/cardboard/cardboard'
cardboard_arr = np.empty(1)
for i in range(1, 404):
    img = load_img(cardboard_str + str(i) + '.jpg')

    # convert to numpy array
    img_array = [img_to_array(img)]

    if i == 1:
        cardboard_arr = np.array(img_array)
    else:
        cardboard_arr = np.append(cardboard_arr,img_array, axis=0)
    print('cardboard' + str(i))
    
glass_str='archive/Garbage classification/Garbage classification/glass/glass'
glass_arr = np.empty(1)
for i in range(1, 502):
    img = load_img(glass_str + str(i) + '.jpg')

    # convert to numpy array
    img_array = [img_to_array(img)]

    if i == 1:
        glass_arr = np.array(img_array)
    else:
        glass_arr = np.append(glass_arr,img_array, axis=0)
    print('glass' + str(i))
    
paper_str='archive/Garbage classification/Garbage classification/paper/paper'
paper_arr = np.empty(1)
for i in range(1, 595):
    img = load_img(paper_str + str(i) + '.jpg')

    # convert to numpy array
    img_array = [img_to_array(img)]

    if i == 1:
        paper_arr = np.array(img_array)
    else:
        paper_arr = np.append(paper_arr,img_array, axis=0)
    print('paper' + str(i))
    
metal_str='archive/Garbage classification/Garbage classification/metal/metal'
metal_arr = np.empty(411)
for i in range(1, 4):
    img = load_img(metal_str + str(i) + '.jpg')

    # convert to numpy array
    img_array = [img_to_array(img)]

    if i == 1:
        metal_arr = np.array(img_array)
    else:
        metal_arr = np.append(metal_arr,img_array, axis=0)
    print('metal' + str(i))
    
plastic_str='archive/Garbage classification/Garbage classification/plastic/plastic'
plastic_arr = np.empty(1)
for i in range(1, 483):
    img = load_img(plastic_str + str(i) + '.jpg')

    # convert to numpy array
    img_array = [img_to_array(img)]

    if i == 1:
        plastic_arr = np.array(img_array)
    else:
        plastic_arr = np.append(plastic_arr,img_array, axis=0)
    print('plastic' + str(i))
    
trash_str='archive/Garbage classification/Garbage classification/trash/trash'
trash_arr = np.empty(1)
for i in range(1, 138):
    img = load_img(trash_str + str(i) + '.jpg')

    # convert to numpy array
    img_array = [img_to_array(img)]

    if i == 1:
        trash_arr = np.array(img_array)
    else:
        trash_arr = np.append(trash_arr,img_array, axis=0)
    print('trash' + str(i))


np.save('cardboard_data', cardboard_arr)
np.save('glass_data', glass_arr)
np.save('metal_data', metal_arr)
np.save('paper_data', paper_arr)
np.save('plastic_data', plastic_arr)
np.save('trash_data', trash_data)
```
