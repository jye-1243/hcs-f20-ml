```python
# import required libraries
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import array_to_img
import math
```

```python
path='chinese mnist/data/data/input_'

arrays = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]

for person in range(1, 101):
    for i in range(1,11):
        for j in range(1,16):
            img = load_img(path + str(person) + '_' + str(i) + '_' + str(j) + '.jpg')
            
            # convert to numpy array
            img_array = img_to_array(img)
            bw_array = np.empty(1)
                
            for x in range(len(img_array)):
                for y in range(len(img_array)):
                    if i == 0 and j == 0:
                        bw_array = np.array(img_array[x][y][0])
                    else:
                        bw_array = np.append(bw_array, img_array[x][y][0])       
            arrays[j-1].append(bw_array)
            
            print(str(person) + '_' + str(i) + '_' + str(j))

print(arrays[0])

np.save('zeros', arrays[0])
np.save('ones', arrays[1])
np.save('twos', arrays[2])
np.save('threes',arrays[3])
np.save('fours',arrays[4])
np.save('fives',arrays[5])
np.save('sixes',arrays[6])
np.save('sevens', arrays[7])
np.save('eights', arrays[8])
np.save('nines',arrays[9])
np.save('tens', arrays[10])
np.save('hundreds',arrays[11])
np.save('thousands',arrays[12])
np.save('tenthous',arrays[13])
np.save('tenmils',arrays[14])
```
